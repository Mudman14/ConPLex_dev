```python
from sklearn.model_selection import ParameterGrid
```

```python
task = ['davis','bindingdb','biosnap']
# task = ['biosnap_prot','biosnap_mol']
model_type = ['SimpleCosine']
mol_feat = ['Morgan_f']
# prot_feat = ['BeplerBerger_f',
#              'BeplerBerger_DSCRIPT_cat_f',
#              'ESM_f',
#              'ESM_DSCRIPT_cat_f',
#              'ProtBert_f',
#              'ProtBert_DSCRIPT_cat_f'
#             ]
prot_feat = ['BeplerBerger_DSCRIPT_f',
             'ESM_DSCRIPT_f',
             'ProtBert_DSCRIPT_f'
            ]
pgrid = ParameterGrid({
    "task": task,
    "model_type": model_type,
    "mol_feat": mol_feat,
    "prot_feat": prot_feat,
})
```

```python
param_sets = []
run_list = []
for p in pgrid:
    param_sets.append(p)

param_sets.sort(key=lambda x: x['task'],reverse=True)
for p in param_sets:
    for rp in range(5):
        run_list.append(f"python train_attendti.py --replicate {rp} --wandb-proj Adapting_PLM_DTI --task {p['task']} --model-type {p['model_type']} --mol-feat {p['mol_feat']} --prot-feat {p['prot_feat']} --exp-id {p['task'].title()}_{p['model_type']}_{p['mol_feat'].rstrip('_f')}_{p['prot_feat'].rstrip('_f')}_rep{rp}")
```

```python
run_list.sort(key = lambda x: x[-1])
print(len(run_list))
for ln in run_list:
    print(ln)
```
